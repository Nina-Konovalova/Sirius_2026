## Что такое EDA (Exploratory Data Analysis)

**EDA (Exploratory Data Analysis)** — это этап работы с данными, на котором мы просто «знакомимся» с датасетом и пытаемся понять, что в нём происходит, без сложных моделей и прогнозов.

Главные цели EDA:
- понять, какие есть признаки (столбцы) и что они означают;
- заметить ошибки, пропуски, аномальные значения;
- увидеть общие закономерности и связи между признаками;
- сформулировать первые гипотезы: «от чего может зависеть целевая переменная?».

Обычно в EDA делают:
- первичный обзор: размер таблицы, типы данных, описательные статистики;
- простые визуализации: гистограммы, боксплоты, графики рассеяния, тепловые карты корреляций;
- проверку предположений: линейность, наличие выбросов, распределения признаков.

Важно, что на этапе EDA мы **исследуем**, а не доказываем теоремы:
- можно задавать много вопросов к данным;
- пробовать разные графики;
- замечать необычные паттерны и идеи для дальнейшего анализа или моделей.

Хороший EDA помогает:
- лучше понять задачу;
- выбрать важные признаки;
- заранее увидеть проблемы, которые могут «сломать» модель (например, сильные выбросы или пропуски);
- сформулировать понятные и осмысленные выводы на человеческом языке.
