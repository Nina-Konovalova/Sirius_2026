## Формализация задачи

Будем строить линейную модель вида:
$$
\hat{y} = w_0 + w_1 x_1 + w_2 x_2,
$$
где:

- $y$ — стоимость аренды
- $x_1$ — площадь по данным агентства
- $x_2$ — полезная площадь
- $w_0, w_1, w_2$ — параметры модели

В матричной записи задача имеет вид:
$$
\hat{y} = Xw,
$$
где $X \in \mathbb{R}^{n \times p}$ — матрица признаков, а $w \in \mathbb{R}^{p}$ — вектор параметров.

Первый столбец матрицы $X$ состоит из единиц и соответствует свободному члену модели.

По классике возьмем mean square error (MSE), или если на русском, то квадрата евклидовой нормы вектора:
$$
MSE(w) = \|Xw - y\|_2^2
$$
Точку минимума можно найти разными способами. Остановимся в этот раз на аналитическом методе. $$
X^T X w = X^T y
$$

При обратимости матрицы $X^T X$ получаем аналитическую формулу:
$$
\hat{w} = (X^T X)^{-1} X^T y
$$
